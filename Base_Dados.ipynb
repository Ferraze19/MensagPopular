```python
# Obter dados do excel
def Criar_lista_doadores ():
    pd.options.display.float_format = "{:,.2f}".format
    lista_doadores = pd.read_excel("Doadores UP.xlsx", sheet_name="Plan1")
    lista_doadores['Valor'] = lista_doadores['Valor'].astype('float64')
    lista_doadores['Numero Recibo'] = lista_doadores['Numero Recibo'].fillna(0.0).astype('int64')
    lista_doadores['CPF'] = lista_doadores['CPF'].fillna(0.0).astype('int64')
    return lista_doadores
```
